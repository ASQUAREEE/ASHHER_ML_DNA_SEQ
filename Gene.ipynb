```python
df_class= pd.read_csv('pp5i_train_class.txt')

df_class
```

```python
class_np= df_class.to_numpy()
```

```python
L=preprocessing.LabelEncoder()
L.fit(class_np)
```

```python
train_class=L.transform(class_np)
train_class
```

```python
L.classes_
```

```python
v_test = v_test.clip(2,16000)
v_train = v_train.clip(2,16000)
```

```python
from sklearn.feature_selection import mutual_info_classif
import matplotlib.pyplot as plt
```

```python
importances = mutual_info_classif(v_train, train_class)
```

```python
importances
```

```python
feat_importances = pd.Series(importances, v_train.columns[0:len(v_train.columns)])
feat_importances
```

```python
imp_gene_df= pd.DataFrame(feat_importances)
```

```python
imp_gene_df.sort_values(0,ascending=False).head(10)
```

```python
new_df = v_train.T[:]
new_df['Rank'] = feat_importances
new_df = new_df.sort_values('Rank',ascending = False)
```

```python
new_df1 = v_test.T[:]
new_df1['Rank'] = feat_importances
new_df1= new_df1.sort_values('Rank',ascending = False)
```
